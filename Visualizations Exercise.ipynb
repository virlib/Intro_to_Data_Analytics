```python
x2= np.arange(0, 5.5, 0.5)    #5.5 not inclusive
plt.plot(x2, (x2)**2, 'r',label='X^2')


plt.plot(x2[:6], x2[:6]**2, 'b--', label='blue x^2')    #6 not inclusive
plt.title('Blue red graph')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.xticks([0,1,2,3,4,5,6])
plt.show()
```

SUBPLOTS EXAMPLE

```python
x = np.arange(-10, 11)
plt.figure(figsize=(12, 6))
plt.title('My Nice Plot')

plt.subplot(1, 2, 1)  # rows, columns, panel selected
plt.plot(x, x ** 2)
plt.plot([0, 0, 0], [-10, 0, 100])
plt.legend(['X^2', 'Vertical Line'])
plt.xlabel('X')
plt.ylabel('X Squared')

plt.subplot(1, 2, 2)
plt.plot(x, -1 * (x ** 2))
plt.plot([-10, 0, 10], [-50, -50, -50])
plt.legend(['-X^2', 'Horizontal Line'])

plt.xlabel('X')
plt.ylabel('X Squared')
```

```python
# prompt: make a variable fifa which stores fifa_data.csv file location which is in data folder of my drive

fifa = '/content/drive/My Drive/Project1_python/data/fifa_data.csv'
fifa_data=pd.read_csv(fifa)
```

```python
bins = [40,50,60,70,80,90,100]

plt.figure(figsize=(8,8))

plt.hist(fifa_data.Overall, bins=bins, color='#fdceee')

plt.xticks(bins)

plt.ylabel('Number of Players')
plt.xlabel('Skill Level')
plt.title('Distribution of Player Skills in FIFA 2018')

plt.show()
```

```python
fifa_data['Preferred Foot'].head()
```

```python
left=(fifa_data['Preferred Foot']=='Left').sum()
right=(fifa_data['Preferred Foot']=='Right').sum()

plt.figure(figsize=(8,5))

labels = ['Left', 'Right']
colors = ['#dbcdef', '#dabbcc']

plt.pie([left, right], labels = labels, colors=colors, autopct='%.3f %%')

plt.title('Foot Preference of FIFA Players')

plt.show()
```

```python
plt.figure(figsize=(8,5), dpi=100)

#plt.style.use('ggplot')

#removing lbs in weight
fifa_data.Weight = [int(x.strip('lbs')) if type(x)==str else x for x in fifa_data.Weight]

light = (fifa_data.Weight < 125).sum()
light_medium = ((fifa_data.Weight >= 125) & (fifa_data.Weight < 150)).sum()
medium = ((fifa_data.Weight >= 150) & (fifa_data.Weight < 175)).sum()
medium_heavy = ((fifa_data.Weight >= 175) & (fifa_data.Weight < 200)).sum()
heavy = (fifa_data.Weight >= 200).sum()

weights = [light,light_medium, medium, medium_heavy, heavy]
label = ['under 125', '125-150', '150-175', '175-200', 'over 200']
explode = (.4,.1,0,0,.4)

plt.title('Weight of Professional Soccer Players (lbs)')

plt.pie(weights, labels=label, explode=explode, pctdistance=0.8,autopct='%.2f %%')
plt.show()
```
